```python
import numpy as np
taxi= np.genfromtxt("nyc_taxis.csv", delimiter=",", skip_header=1)
```

```python
speed=taxi[:,7]/(taxi[:,8]/3600)
#speed in hrs
meanspeed=speed.mean()
print("Mean speed =",meanspeed)
```

```python
febride=taxi[taxi[:,1]==2]
print("Rides taken in feb = ",len(febride))
```

```python
tip=taxi[taxi[:,12]>50]
print("Rides where tip was more than 50 dollars = ",len(tip))
```

```python
jfk=taxi[taxi[:,6]==2]
print("Rides where dropoff location is jfk airport = ",len(jfk))
```
